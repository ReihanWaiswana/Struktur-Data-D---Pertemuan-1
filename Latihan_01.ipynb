1.5 Nama Mahasiswa

```python
mahasiswa = ["Andi", "Budi", "Ega", "Rehan", "Gusde"]
print("List Mahasiswa: ")
print(mahasiswa)
```

Tambahan 1 Nama Baru dengan append

```python
mahasiswa.append("Anomali")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus Satu Elemen Menggunakan Remove

```python
mahasiswa.remove("Anomali")
print("Setelah menghapus nama Anomali")
print(mahasiswa)
```
